## Pregunta 1

````python
df_datosxpoblacion = df_datos_comuna.merge(df_poblacion, left_on="id", right_on="CODIGO")
df_datosxpoblacion
````

````python
df_datosxpoblacion = df_datosxpoblacion.dropna()
df_datosxpoblacion
````

````python
df_datosxpoblacion = df_datosxpoblacion[["MUNICIPIO","POBLACIÓN","metros_plaza"]]
df_datosxpoblacion
````

````python
df_datosxpoblacion.columns= ["Comuna","Poblacion","Metros_de_plaza"]
df_datosxpoblacion
````

````python
df_datosxpoblacion = df_datosxpoblacion.assign(MdP_P=lambda x: x.Metros_de_plaza/x.Poblacion)
df_datosxpoblacion
````

````python
df_datosxpoblacion = df_datosxpoblacion[["Comuna","MdP_P"]]
df_datosxpoblacion
````

````python
df_datosxpoblacion = df_datosxpoblacion.sort_values("MdP_P", ascending= False)
df_datosxpoblacion
````

Obtenida esta tabla se puede decir que las 3 comunas que poseen una mayor proporción entre los metros de plaza y la población son: Rinconada con 3451.166, Primavera con 24.832 y Río Ibañez con 17.717. Y las comunas con una menor proporción son: Lautaro con 0.002, Nogales con 0.005 y Andacollo con 0.006.

## Pregunta 2

````python
df_presupuesto = df_presupuesto.replace('No Recepcionado', 0)
df_presupuesto["PRESUPUESTO"] = df_presupuesto["PRESUPUESTO"].astype(int)
df_presupuesto.info()
````

````python
df_presxcod = df_presupuesto.merge(df_codigos, left_on="CODIGO", right_on="Código Comuna 2018")
df_presxcod
````

````python
df_presxcod = df_presxcod[["Nombre Región","PRESUPUESTO"]]
df_presxcod
````

````python
df_presxcod.columns = ["Region","Presupuesto"]
df_presxcod
````

````python
df_presxcod["Presupuesto"].groupby([df_presxcod['Region']]).mean().sort_values(ascending=False)
````

Ya habiendo ejecutado los codigos y obtener la presente tabla que contiene el presupuesto promedio de cada región Chilena ordenados de mayor a menor. Se puede concluir que la región que tiene el mayor presupuesto promedio es la Región Metropolitana de Santiago con $43.416.560.
