## Pregunta 1

````python
df_presupuesto=df_presupuesto.replace("No Recepcionado",0)
df_presupuesto["PRESUPUESTO"]=df_presupuesto["PRESUPUESTO"].astype(int)
df_presupuesto.info()
````

````python
df_presupuestocodigo=df_presupuesto.merge(df_codigos,left_on="CODIGO",right_on="Código Comuna 2018")
df_presupuestocodigo
````

````python
df_presupuestocodigo=df_presupuestocodigo[["Nombre Provincia","PRESUPUESTO"]]
df_presupuestocodigo
````

````python
df_presupuestocodigo.columns=["Provincia","Presupuesto"]
df_presupuestocodigo["Presupuesto"].groupby([df_presupuestocodigo["Provincia"]]).mean().sort_values(ascending=False)
````

Se puede concluir a partir de los resultados que la provincia que tiene mayor presupuesto promedio es la de Santiago, esto puede ser debido a que es la Provincia con mayor cantidad de habitantes en el pais.
